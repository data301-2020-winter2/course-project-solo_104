## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from scripts import project_functions as pf

df = pf.load_and_process('../data/raw/dataset.csv')

pf.general_info(df)
```

This provides us with a **general description** of the dataset, along with some other information.

```python
pf.streaks(df)
```

This shows us that win streaks as long as **8 or 9 years** have existed, meaning there has been some very dominant dynasties in the past.

```python
pf.placements(df,'First Place',10)
```

This plot gives a general idea of who the most **dominant teams** are throughout history, by showing the teams with the most championships under their belt.

```python
pf.placements(df,'Second Place',10)
```

This graphs tell us what teams just fall short *(place 2nd)* the most.

```python
sns.countplot(y='Year',data=df,hue='Win Streak',order=df['Year'].value_counts().index[:20])
```

This plot shows the **most occurring years** in the dataset, meaning the most sport championships took place in these top 20 years, while other years may have less because of event cancellations or missing data in the dataset. It is interesting that some years, like 2010 shown above, had **no** 1 win streaks, meaning in all sport championships that year, **no** new team came out on top *(all First Place finishers had won the last year as well)*.
