## Research Question: *What are the most dominant sports dynasties throughout history?*

To find some of the most dominant teams through history, we can start by recalling the count plot we discovered in out EDA, that gave us the most winning sports franchises throughout history, but expand the plot to include 20 teams:

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from scripts import project_functions as pf

df = pf.load_and_process('../data/raw/dataset.csv')

pf.placements(df,'First Place',20)
```

This gives us the teams with the most first place finishes, but to find out which dynasties were most dominant during a certain time, we may want to see which of these 20 teams have impressive win streaks, instead of just a bunch of 1 win streaks spread throughout the team's existence.

```python
pf.placements_and_streaks(df,'First Place',20)
```

Just in case, plot the sport of the top 20 most winning teams, so we know what sport we are talking about in particular:

```python
pf.topteam_sport(df)
```

We can analyze the amount of First Place finishes, along with the teams with higher win streaks, to get a generally good idea of who the most dominant dynasties may be. I found that the top 10 best teams, based on the plot above, seemed to be the following:
- North Carolina Tar Heels (Womens College Soccer)
- New York Yankees (MLB)
- Montreal Canadiens (NHL)
- USC Trojans (Mens College Baseball)
- UCLA Bruins (Mens College Basketball)
- Yale Bulldogs (Mens College Football)
- Boston Celtics (NBA)
- Connecticut Huskies (Womens College Basketball)
- Edmonton Eskimos (CFL)
- Green Bay Packers (NFL)

*(I went off of highest win streaks first, then most wins, because I believe a power team is truly shown by how many in a row they win.)*

Since each bar on the plot equals 1 year where they won, and touching bars (the wider-looking bars) means they won consecutive years, we can determine the specific top 10 sports dynasties based on these plots:
- 1982-2012 Tar Heels (Womens College Soccer)
- 1936-1963 Yankees (MLB)
- 1953-1980 Canadiens (NHL)
- 1968-1979 Trojans (Mens College Baseball)
- 1964-1976 Bruins (Mens College Basketball)
- 1875-1896 Bulldogs (Mens College Football)
- 1957-1970 Celtics (NBA)
- 1995-2016 Huskies (Womens College Basketball)
- 1975-1983 Eskimos (CFL)
- 1961-1968 Packers (NFL)
*(As a rule, I decided a gap of more than 5 years between wins meant the end of a dynasty.)*
